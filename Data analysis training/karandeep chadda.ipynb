# ___CLEANING THE GIVEN DATASET___

```python
# I HAVE ALREADY CLEANED IT SO DONT RUN IT AGAIN!!!!!!!

# the given data set has non alphabetical-characters.we have to replace them!

import pandas as pd
import unicodedata

# Load the dataset
df = pd.read_csv('Dataset.csv')

# Removing cells Missing Values
df.dropna(inplace=True)

# Removing Duplicates
df.drop_duplicates(inplace=True)

# Function to replace non-ASCII characters with ASCII equivalents
def replace_non_ascii(text):
    return ''.join(char if ord(char) < 128 else unicodedata.normalize('NFKD', char).encode('ascii', 'ignore').decode('utf-8', 'ignore') for char in str(text))

# Replace non-ASCII characters in all columns
df = df.map(replace_non_ascii)

# Reset index after modifying DataFrame
df.reset_index(drop=True, inplace=True)

# (I HAVE SAVED IT ALREADY! THATS WHY I HAVE COMMENTED THE BELOW CODE!)

# df.to_csv('cleaned_dataset.csv', index=False) 
```

# ___TASK 4: Restaurant Chains___

* ### ___TASK 4.1___
"Identify if there are any restaurant chains present in the dataset"

```python
import pandas as pd

df = pd.read_csv('cleaned_dataset.csv')

restaurant_names = df['Restaurant Name'].str.lower()

# a dictionary to store potential restaurant chains
restaurant_chains = {}

# Iterate over each restaurant name
for name in restaurant_names:
    for chain_name in restaurant_chains.keys():
        if chain_name in name:
            restaurant_chains[chain_name].append(name)
            break
    else:
        restaurant_chains[name] = [name]

potential_chains = [(chain_name, restaurants) for chain_name, restaurants in restaurant_chains.items() if len(restaurants) > 1]

# Print potential restaurant chains

print("Potential restaurant chains are:-")
print()
for chain_name, restaurants in potential_chains:
    print(f"Chain: {chain_name.capitalize()}")
    for restaurant in restaurants:
        print(f"-{restaurant.title()}")
    print("....................")
    print()
```

* ### ___TASK 4.2___
"Analyze the ratings and popularity of different restaurant chains"

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('cleaned_dataset.csv')

# Aggregate data by restaurant chains
chain_stats = df.groupby('Restaurant Name').agg({'Aggregate rating': 'mean', 'Votes': 'sum', 'City': 'count'})

# Rename columns for clarity
chain_stats.rename(columns={'Aggregate rating': 'Average Rating', 'Votes':  'Total Votes', 'City':  'Total Restaurants'}, inplace=True)

# Sort by total votes in descending order
chain_stats = chain_stats.sort_values(by='Total Votes', ascending=False)

# Print the summary statistics for the top 10 restaurant chains
print("Summary for the top 10 Restaurant chains:\n")
print(chain_stats.head(10))
```
